# Extração de Dados pelo Arduino
A extração de dados deverá ser feita via porta COM, onde os dados enviados para o monitor serial no Arduino são extraídos pelo comunicação serial do programa abaixo. Os dados chegam sem pré-processamento adequado. Portanto, é feito um pré-processamento para transformar em arquivo csv. Lembre-se que esse processo é apenas para simulação do funcionamento do sistema.
## Orientações
O Monitor serial deve estar fechado na Arduino IDE. Resete o Arduino com o monitor serial fechado, e logo em seguida, execute o código python. O programa deverá coletar todos os dados corretamente. Talvez seja necessário desconectar o arduino da porta COM e conectar novamente, logo em seguida executando o programa para a aquisição de dados.

# Importação de Bibliotecas

```python
import serial
import csv
import time
```

```python
executando = True
contador = 0
linhas_salvas = 0

print("Iniciando comunicação serial...")

porta_serial = serial.Serial('COM5', 9600, timeout=1)
nome_arquivo = '../Data/dados3.csv'

print(f"Abrindo arquivo '{nome_arquivo}' para gravação...")

with open(nome_arquivo, mode='w', newline='') as arquivo_csv:
    escritor_csv = csv.writer(arquivo_csv)
    print("Arquivo aberto. Iniciando captura de dados...\n")

    while executando:
        linha = porta_serial.readline().decode(errors='ignore').strip()
        contador += 1

        if linha:
            valores = linha.split(',')

            if len(valores) >= 27:
                escritor_csv.writerow(valores)
                linhas_salvas += 1
                print(f"[{contador}] DADO SALVO ({len(valores)} colunas)")
            else:
                print(f"[{contador}] LINHA DESCARTADA ({len(valores)} colunas)")

        else:
            print(f"[{contador}] Nenhum dado recebido...")

        if contador > 300:  # ajuste conforme necessário
            executando = False
            print("\nEncerrando captura de dados...")
            print(f"Total de linhas salvas: {linhas_salvas}")

        time.sleep(0.2)

print("Programa finalizado.")
```
